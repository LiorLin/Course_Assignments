```python
import pandas as pd

# Read 'patient_data.csv' into a pandas DataFrame called 'data'
data = pd.read_csv('patient_data.csv') 
print(f"The first five rows of the table: \n{data.head(5)}") # Display the first 5 rows of the dataset

# Count the number of patients : 
patient_count = len(data) 
print(f"\nThere is data about {patient_count} patients in the table.") 
```

```python
# Calculate basic statistics

mean_wbc = data['WBC_Count'].mean()
print(f"The mean white blood count of the patients is {mean_wbc}.")

min_wbc = data['WBC_Count'].min()
print(f"The lowest white blood count is {min_wbc}.")

max_wbc = data['WBC_Count'].max()
print(f"The highest white blood count is {max_wbc}.")

unique_conditions = data['WBC_Count'].unique()
unique_conditions_count = len(unique_conditions)

if unique_conditions_count == patient_count:
    print (f"\nEach white blood count appears once.")
else:
    print(f"\nNumber of unique conditions: {unique_conditions_count}")
```

```python
# Filter patients with WBC count > 10,000

high_wbc = data[data['WBC_Count'] > 10000] # Creating a new dataset that is filtered based on WBC_Count values. 
# High_wbc is a dictionary containing the filtered patientID numbers and their indexes in the original dataset. 

print(f"Pateints with white blood cell count higher than 10,000: {high_wbc.PatientID.values}")
print(high_wbc)

# Filter patients with the condition 'Healthy'

healthy_patients = data[data['Condition'] == 'Healthy']

print(f"\nPateints with the condition 'Healthy': {healthy_patients.PatientID.values}")
print(healthy_patients)
```

```python
# Count patients by condition

conditions = data['Condition']
# print(conditions) - for convinience

condition_counts = {'Healthy': 0, 'Infection': 0, 'Cancer': 0} #Initiate a condition counter dictionary.

for condition in conditions:
    condition_counts[condition] +=1

for key, value in condition_counts.items():
    print(f"There are {value} {key} patients.")

print(f"\nNumber of patients per condition: {condition_counts}")
    
    
    
```

```python
import matplotlib.pyplot as plt

# Bar chart for number of patients by condition
plt.bar(list(condition_counts.keys()), list(condition_counts.values()))

# Customizing the bar plot : 

plt.title('Number of patients per condition')
plt.ylabel('Number of patients')

# Display the graph 
plt.show()


# Scatter plot of Age vs. WBC Count
plt.scatter(data['Age'], data['WBC_Count'])

# Customizing the scatter plot : 

plt.title('Whites blood cell count with age')
plt.xlabel('Age')
plt.ylabel('White blood cell count')

# Display the graph 
plt.show()

            
```
