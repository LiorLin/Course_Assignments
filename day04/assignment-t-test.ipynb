```python
import numpy as np

# Generate random distributions
dist1 = np.random.randn(100)
dist2 = dist1+0.3
```

```python
from scipy.stats import ttest_ind

# Performing t-test on the 2 populations using the function `ttest_ind()`:
t_stat, p_value = ttest_ind(dist1, dist2)

# Printing the t-statistic and p-value, and whether significant:  
print(f"t-statistic: {t_stat}")
if p_value < 0.05:
    print (f"The two populations are significantly different from each other as p-value= {p_value:.2f}")
else:
    print(f"The two distributions are not significantly different as p-value= {p_value:.2f}")
 
```

```python
import matplotlib.pyplot as plt

# Plot histograms
plt.hist(dist1, color='c', label = 'Weizmann') 
plt.hist(dist2, color = 'm', label = 'Tau')

plt.title('Refrigerator\'s Temperature distributions')
plt.xlabel('Serial number')
plt.ylabel('Temperature')
plt.legend() 

plt.show()
```

```python
# Create boxplots
plt.boxplot([dist1, dist2])#TODO:
plt.title('Refrigerator\'s Temperature in Boxplot')
plt.xticks([1, 2], ['Weizmann', 'TAU'])
plt.ylabel('Temperature')
plt.show()
```
